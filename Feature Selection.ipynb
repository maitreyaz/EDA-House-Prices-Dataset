```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.pandas.set_option('display.max_columns', None)

dataset = pd.read_csv('X_train.csv')
dataset.head()
```

```python
# ytrain = dataset['SalePrice'] ytrain is a list heare
ytrain = dataset[['SalePrice']] #ytrain is a dataframe here (check the extra square braces)
xtrain = dataset.drop(['Id', 'SalePrice'], axis=1)
xtrain.head()
```

```python
ytrain.head()
```

```python
#Appication of SelectFromModel after specification of Lasso regr model with proper alpha and random state values:
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(xtrain, ytrain)
```

```python
feature_sel_model.get_support() #True if feature imp, else false
```

```python
selected_features = xtrain.columns[feature_sel_model.get_support()]
selected_features
```

```python
print("Total Features : {}".format(len(xtrain.columns)))
print("No. of Selected Features : {}".format(len(selected_features)))
```

```python
xtrain = xtrain[selected_features]
xtrain.head()
```
